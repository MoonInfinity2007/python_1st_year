```python
def fetch_lyrics_page(url):
    responds = requests.get(url)
    if not responds:
        return []
    soup = BeautifulSoup(responds.text, 'html.parser')
    h_text = soup.find("div", attrs={"class": "Lyrics__Container-sc-c1895f55-1 fYBzEj"})
    header = h_text.find("div", attrs={"class": "LyricsHeader__Container-sc-347d044d-1 gPVEFK"})
    if (header): header.decompose()
    return [elem.text for elem in h_text]
```

```python
import json
def get_lyrics(response):
    result = json.loads(response.text)
    if res := result.get('response').get('hits'):
        return fetch_lyrics_page("https://genius.com/"+res[0]["result"]["path"])
    return []
```

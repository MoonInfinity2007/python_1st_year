### 1 TASK

```python
import csv
data = []
mp = dict()
with open('dataset.csv', 'r', newline='') as csvfile:
    data_csv = csv.reader(csvfile, delimiter=',')
    iki = 0
    for row in data_csv:
        iki += 1
        if (iki == 1):
            row0 = row 
            continue
        tp = (row[2], row[4])
        if not(tp in mp): data.append(row)
        mp[tp] = 1
len(data)
```

### 2 TASK

```python
dp = dict()
for i in range(len(data)):
    artists = data[i][2].split(';')
    artists = set(artists)
    for artist in artists:
        if not(artist in dp): dp[artist] = 0
        dp[artist] += 1
```

```python
sdp = sorted(dp.items(), key=lambda x: x[1], reverse=True)
sdp[:10]
```

### 3 TASK

```python
popularity_dict = dict()
C = 0
for i in range(len(data)):
    popularity = int(data[i][5])
    C += popularity
    artists = data[i][2].split(';')
    for artist in artists:
        if not(artist in popularity_dict): popularity_dict[artist] = []
        popularity_dict[artist].append(popularity)
C /= len(data)
C
```

```python
hit_makers = dict()
for elem in popularity_dict:
    spd = sorted(popularity_dict[elem])
    hit_makers[elem] = (len(spd) / (len(spd) + 5)) * sum(spd) / len(spd) + (5 / (len(spd) + 5)) * C
```

```python
shit_makers = sorted(hit_makers.items(), key=lambda x: x[1], reverse=True)
shit_makers[:10]
```

### 4 TASK

```python
import re
def split_artists(artist_string):
    parts = re.split(r'\s*(?:,|&|feat\.|ft\.|featuring)\s*', artist_string, flags=re.IGNORECASE)
    return [a.strip(' () ') for a in parts if a.strip(' () ')]
```

```python
feat_artist = dict()
for i in range(len(data)):
    track_name = data[i][4]
    ind = track_name.find('feat')
    if (ind != -1):
        st = split_artists(track_name[ind + 6:])
        st = set(st)
        for artist in st:
            if not(artist in feat_artist): feat_artist[artist] = 0
            feat_artist[artist] += 1
len(feat_artist)
```
